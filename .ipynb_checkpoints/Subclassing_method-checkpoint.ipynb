```python
import numpy as np
import pandas as pd 
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
dir(data)
```

```python
feature=np.array(data.data)
feature[:10]
```

```python
data.target_names
```

```python
target=np.array(data.target)
target
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(feature,target,test_size=0.2,random_state=42,stratify=target)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, optimizers
```

```python
class IrisNet(tf.keras.Model):
    def __init__(self):
        super().__init__()
        self.d1 = layers.Dense(32, activation='relu')
        self.d2 = layers.Dense(12, activation='relu')
        self.out = layers.Dense(3, activation='softmax')

    def call(self, inputs, training=False):
        x = self.d1(inputs)
        x = self.d2(x)
        return self.out(x)
```

```python
model = IrisNet()
```

```python
model.build(input_shape=(4 ))
```

```python
model.compile(optimizer=optimizers.Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train
history = model.fit(xtrain, ytrain, validation_split=0.2, epochs=50, batch_size=8, verbose=1)
```

```python
model.summary()
```

```python
y_pred = model.predict(xtest)
```

```python
y_pred
```

```python
y_pred = y_pred.argmax(axis=-1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_pred)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
